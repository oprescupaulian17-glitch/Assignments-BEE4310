    To get started using ZeroInflatedPoisson, follow the steps below:
    - subset the ebird_habitat dataframe into a new dataframe (called X_exog) using only the columns from ebird_habitat that correspond to the predictors for the Poisson process (listed above). 
        - Convert X_exog into a numpy ndarray using X_exog = X_exog.values

    - Repeat the step above, but this time create the matrix X_infl for the predictors to be used in the inflationary process. 
    - Next, create a variable y equal to the 'observation_count' column of ebird_habitat
        - To ensure y is a numpy array with a single column, set y = ebird_habitat['observation_count'].values.reshape(-1,1)
        - y will be used for the 'endog' argument in ZeroInflatedPoisson  
    
    Now define a ZIP regression object called ZIP_model and fit the model. You may need to increase maxiter in the fit() method to ensure the algorithm converges (e.g., ZIP_model.fit(maxiter=1000)). Look up the documentation for ZeroInflatedPoisson from statsmodels.discrete.count_model for more detail on how to use this function. 

    **Once the model has fit, report its AIC, as well as the AIC for the original Poisson regression from problem #7. Based on the AIC values, which model would you choose and why?**  
    
    **Finally, report the regression table from the ZIP regression model using the summary command. Comment on the statistical significance of the elev_sd predictor in the original Poisson regression and the ZIP. Have your conclusions changed about the effect of this predictor on thrush counts?** 
    - Note: In the ZIP, we did not use a formula when fitting the model, but instead used numpy arrays X_exog, X_infl, and y as arguments. This is because there is no formula support for zero-inflated Poisson regression in statsmodels at this time. Therefore, the predictors are unamed in the summary table. However, in the summary table, the predictors for the Poisson process and the inflationary process are listed in the order they were provided in X_exog and X_infl, which will allow you to answer this question.

```python
#9
# Define predictors for Poisson Counting Process
predictors = ['day_of_year', 'elevation_median', 'elevation_sd', 'pland_forest', 'pland_grassland_shrubland_savanna', 'pland_wetland', 'pland_cropland', 'pland_urban', 'pland_other']
X_exog = ebird_habitat[predictors]
X_exog = X_exog.values
X_exog = sm.add_constant(X_exog)

# Define predictors for excess zero process
zero_inflation_predictors = ['time_observations_started', 'duration_minutes', 'effort_distance_km', 'number_observers']
X_infl = ebird_habitat[zero_inflation_predictors]
X_infl = X_infl.values
X_infl = sm.add_constant(X_infl)

# Degine response variable
y = ebird_habitat['observation_count'].values.reshape(-1,1)

# Fit Zero-Inflated Poisson Model
ZIP_model = ZeroInflatedPoisson(endog=y, exog=X_exog, exog_infl=X_infl, inflation='logit').fit(maxiter=1000, disp=0)

# Report and compare AIC values
print("Poisson GLM AIC:", model_poisson.aic)
print("Zero-Inflated Poisson Model AIC:", ZIP_model.aic)
if ZIP_model.aic < model_poisson.aic:
    print("The Zero-Inflated Poisson model provides a better fit based on AIC.")
else:
    print("The Poisson GLM provides a better fit based on AIC.")

# Print summary of ZIP model
print(ZIP_model.summary())
```

The smaller AIC score showcases the better model, which is the ZIP.

The elevation_sd predictor appaers to be more significant in the original Poisson regression, with P>|z| being 0.035, whilst in the ZIP, x3 has a significance level of 0.118. The conclusions change in case we would approach the hypothetical testing with a conventional alpha of 0.05, hereby in the original model proving significant, whilst in the ZIP not.
